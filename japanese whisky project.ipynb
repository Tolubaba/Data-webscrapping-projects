```python
import requests
from bs4 import BeautifulSoup
import time
import smtplib
import csv
import pandas as pd
```

```python
#we craete two empty array profuctlinks and whsikylists
productlinks=[]
whiskylist=[]
#using the for range because we want to iterate only the first and second page to get data from
for x in range(1,2):
    url=('https://www.thewhiskyexchange.com/c/35/japanese-whisky')
    headers={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.102"}
    baseurl=('https://www.thewhiskyexchange.com/')
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'lxml')
    soup2=BeautifulSoup(soup.prettify(),'lxml')
    #telling the complier to find all the li in whcih has class of product-grid__item
    product= soup2.find_all('li', class_="product-grid__item")
    #by using the for loop we want tpo iterate throught the product
    for item in product:
        #telling the complierr to look for all a attribute or tag in the html
        #the use of  href=true implies its a link
        for link in item.find_all('a',href=True):
            #to get the specific inforation of each item we added the base url to the link gooten from tlooping tru item so can can directy have acess to the rach content
            productlinks.append(baseurl +link['href'])
            
```

```python
#the end result of addding the base url to the a link
for link in productlinks:
    print(link)
```

```python
def itemlist():
    for link in productlinks:
        r=requests.get(link,headers=headers)
        soup2=BeautifulSoup(r.content,'lxml')
        
        try:
            rating=soup2.find('div',class_='review-overview').text.strip()
            name=soup2.find('h1',class_='product-main__name').text.strip()
            price=soup2.find('p',class_='product-action__price').text.strip()[1:]
        except:
            rating='no rating'
            name='no name'
            price='no price'
           
            
       
        whisky={'name':name,'price':price,'rating':rating,}
        whiskylist.append(whisky)
        return pd.DataFrame(whiskylist)
```

```python
df=itemlist()
```
