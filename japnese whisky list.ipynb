```python
import requests
from bs4 import BeautifulSoup
import time
import smtplib
import csv
import pandas as pd
```

```python
productlinks=[]
whiskylist=[]
for x in range(1,2):
    url=('https://www.thewhiskyexchange.com/c/35/japanese-whisky')
    headers={"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/98.0.4758.102"}
    baseurl=('https://www.thewhiskyexchange.com/')
    page=requests.get(url)
    soup=BeautifulSoup(page.content,'lxml')
    soup2=BeautifulSoup(soup.prettify(),'lxml')
    product= soup2.find_all('li', class_="product-grid__item")
    for item in product:
        for link in item.find_all('a',href=True):
            productlinks.append(baseurl +link['href'])
            
print(productlinks)

    


    
```

```python
for link in productlinks:
    print(link)
```

```python
 for link in productlinks:
        r=requests.get(link,headers=headers)
        soup2=BeautifulSoup(r.content,'lxml')
        
        try:
            rating=soup2.find('div',class_='review-overview').text.strip()
            name=soup2.find('h1',class_='product-main__name').text.strip()
            price=soup2.find('p',class_='product-action__price').text.strip()[1:]
        except:
            rating='no rating'
            name='no name'
            price='no price'
           
            
       
        whisky={'name':name,'price':price,'rating':rating,}
        whiskylist.append(whisky)
        
        print(whisky)
        


#soup4=BeautifulSoup(soup3.prettify(),'lxml')
```

```python

import pandas as pd
df=pd.DataFrame(whiskylist)
df
```

```python
df.to_csv('japanesewhiskylist.csv',index=False)
```
